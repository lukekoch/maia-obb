# YOLOv11 training

```python
from ultralytics import YOLO
from ultralytics.data.split import autosplit

autosplit(
    path="data/in-bucket/images",
    annotated_only=True,
    weights=(0.7, 0.3, 0.0),  # (train, validation, test) fractional splits
)
```

```python
model = YOLO("models/yolo11n-obb.pt")
```

```python
trained_model = model.train(
    data="data/in-bucket/objects.yaml",
    epochs=50,
    mosaic=True,
    degrees=180,
    translate=0.4,
    scale = 0.5,
    hsv_s=0.25,
    hsv_v=0.25
)
```
