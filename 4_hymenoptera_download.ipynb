# Hymenoptera screening dataset download

```python
import requests
import os
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv("hymenoptera-screen/in-export/observations-648956.csv")
df_limited = (
    df.groupby("taxon_id", group_keys=False)
      .apply(lambda g: g.sample(min(len(g), 10), random_state=42))
)
df_limited.to_csv("hymenoptera-screen/in-export/observations-648956_filtered.csv")
display(df_limited)
```

```python
IMG_PER_SPECIES = 10
DATA_DIR = "hymenoptera-screen/images"
```

```python
tax_ids = df_limited["taxon_id"].unique()
for species_id in tqdm(tax_ids):
    observations = df_limited.loc[df_limited["taxon_id"] == species_id]
    for index, observation in observations.iterrows():
        taxon_name = observation["scientific_name"]
        if not os.path.exists(f"{DATA_DIR}/{taxon_name}"):
            os.makedirs(f"{DATA_DIR}/{taxon_name}")
        img_url = f"{observation['image_url']}"
        try:
            img_data = requests.get(img_url).content
            with open(f"{DATA_DIR}/{taxon_name}/{observation['id']}.jpg", 'wb') as f:
                f.write(img_data)
        except Exception as e:
            print(e)
            print(species_id)
```
